```python
# Import the required libraries and modules that you would need
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Read that data into Python and call the dataframe `churnData`
churn_data = pd.read_csv('Customer-Churn.csv')
churn_data.head()
```

```python
churn_data.columns = churn_data.columns.str.replace(' ','_')
churn_data.columns= churn_data.columns.str.lower()
```

```python
churn_data.dtypes
```

```python
# Check the datatypes of all the columns in the data. 
# You would see that the column `TotalCharges` is object type. 
# Convert this column into numeric type using `pd.to_numeric` function.
```

```python
churn_data.totalcharges.isnull().sum()
```

```python
churn_data['totalcharges'] = churn_data['totalcharges'].replace(' ', '0')
```

```python
churn_data['totalcharges'] = pd.to_numeric(churn_data.totalcharges, downcast="float")
```

```python
churn_data.dtypes
```

```python
# Check for null values in the dataframe. 
# Replace the null values.
churn_data.isnull().values.any()
```

```python
churn_data.isnull().sum().sum()
```

```python
# Use the following features: `tenure`, `SeniorCitizen`, `MonthlyCharges` and `TotalCharges`:
#   - Scale the features either by using normalizer or a standard scaler.
#   - Split the data into a training set and a test set.
#   - Fit a logistic regression model on the training data.
#   - Check the accuracy on the test data.
```

```python
features = churn_data[['tenure', 'seniorcitizen', 'monthlycharges', 'totalcharges']]
features
```

```python
transformer = StandardScaler().fit(features)
features_stand = transformer.transform(features)
```

```python
features_stand
```

```python
# Split the data into a training set and a test set.
```

```python
label = churn_data[['churn']]
label
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_stand, label, test_size=0.25, random_state=50)
```

```python
classification = LogisticRegression(random_state=0, solver='lbfgs')
classification.fit(X_train, y_train)
```

```python
print("test data accuracy was ",classification.score(X_test,y_test))
# 100% on training data
print("train data accuracy was ",classification.score(X_train,y_train))
```

```python
# Managing imbalance in the dataset

# - Check for the imbalance.
# - Use the resampling strategies used in class for upsampling and downsampling to create a balance between the two classes.
# - Each time fit the model and see how the accuracy of the model is.
```

```python
label.value_counts()
```

```python
churn_data2 = pd.concat([features, label], axis=1)
churn_data2
```

```python
from sklearn.utils import resample

category_0 = churn_data2[churn_data2['churn'] == 'No']
category_1 = churn_data2[churn_data2['churn'] == 'Yes']
```

```python
# Downsampling

category_0_undersampled = resample(category_0, 
                                   replace=False, 
                                   n_samples = len(category_1))
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
churn_data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
```

```python
churn_data_downsampled['churn'].value_counts()
```

```python
y = churn_data_downsampled[['churn']]
X = churn_data_downsampled.drop(['churn'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=50)
```

```python
classification = LogisticRegression(random_state=0, solver='lbfgs')
classification.fit(X_train, y_train)
```

```python
print("test data accuracy was ",classification.score(X_test,y_test))
# 100% on training data
print("train data accuracy was ",classification.score(X_train,y_train))
```

```python
# Upsampling

category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1_oversampled.shape)
```

```python
churn_data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
churn_data_upsampled['churn'].value_counts()
```

```python
y = churn_data_upsampled[['churn']]
X = churn_data_upsampled.drop(['churn'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=50)
```

```python
classification = LogisticRegression(random_state=0, solver='lbfgs')
classification.fit(X_train, y_train)
```

```python
print("test data accuracy was ",classification.score(X_test,y_test))
# 100% on training data
print("train data accuracy was ",classification.score(X_train,y_train))
```
